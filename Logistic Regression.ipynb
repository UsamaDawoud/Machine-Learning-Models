```python
# the example for hearing test to determine whether it's ok or not according to accepting level
```

```python
df=pd.read_csv('hearing_test.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
X= df.drop('test_result',axis=1)
```

```python
y=df['test_result']
```

```python
# To increase the accuracy a Polynomial Features used
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly=PolynomialFeatures(include_bias=False , degree =3 )
```

```python
pX = poly.fit_transform(X)
```

```python
X.shape
```

```python
pX.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(pX,y,test_size=0.3,random_state=101)
```

```python
# Scalling the data 
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaler.fit(X_train)
```

```python
X_train=scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(solver='saga')
# by default solver ’lbfgs’ is setted , ‘liblinear’ which is for small datasets is a good choice
# ‘sag’ and ‘saga’ are faster for large ones.
# ‘lbfgs’, ‘sag’ and ‘saga’ handle L2 or no penalty
# ‘liblinear’ and ‘saga’ also handle L1 penalty
# ‘saga’ also supports ‘elasticnet’ penalty
# 
# ( L1 is lasso )( L2 is ridge  ) ( elasticnet) : are all belong to Linear regression model
```

```python
# Grid Search to imporve the model 
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {'penalty':['l1','l2','elasticnet'], 'l1_ratio':np.linspace(0,1,20)}
# already changed solver to ‘saga’  to support ['l1','l2','elasticnet'] for panalty ( modifying the model parameters)
# penalty by default is 'L2'
#
# l1_ratio 
# The Elastic-Net mixing parameter, with 0 <= l1_ratio <= 1. Only used if penalty='elasticnet'.
# Setting l1_ratio=0 is equivalent to using penalty='l2',
# while setting l1_ratio=1 is equivalent to using penalty='l1'. 
# For 0 < l1_ratio <1, the penalty is a combination of L1 and L2
```

```python
gridmodel = GridSearchCV(model , param_grid=param_grid)
```

```python
from sklearn.metrics import  accuracy_score , confusion_matrix , classification_report
```

```python
gridmodel.fit(X_train,y_train)
```

```python
y_pred = gridmodel.predict(X_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
# there is 27 false positive & 78 false negitive 
```

```python
print(classification_report(y_test,y_pred))
```
