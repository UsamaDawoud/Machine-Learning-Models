# Grid Search technic is used to figure out the best parameters to be used from a range in model 

```python
import numpy as np
import pandas as pd
```

```python
df= pd.read_csv('Advertising.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
X = df.drop('sales',axis=1)
y=df['sales']
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly = PolynomialFeatures(degree=4,include_bias=False)
```

```python
pX = poly.fit_transform(X)
```

```python
pX.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split (pX,y,test_size=0.3 , random_state=101)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
pX_train = scaler.transform(X_train)
pX_test = scaler.transform(X_test)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.linear_model import ElasticNet
```

```python
model =ElasticNet()
```

```python
param_grid = { 'alpha':[0.1,1,5,10,50,100],
               'l1_ratio':[0.1,0.5,0.9,0.95,0.99,1]}
```

```python
grid = GridSearchCV(estimator= model,
                    param_grid=param_grid,
                    scoring='neg_mean_squared_error',
                    cv =5,
                   verbose=2)
# verbose is an option allows information messages 
```

```python
grid.fit(pX_train,y_train)
```

```python
grid.best_estimator_
```

```python
grid.best_params_
# same as before but it will show in dictionary form
```

```python
pd.DataFrame(grid.cv_results_)
```

```python
y_pred = grid.predict(pX_test)
```

```python
from sklearn.metrics import mean_squared_error , mean_absolute_error
```

```python
np.sqrt(mean_squared_error(y_test,y_pred))
```

```python
mean_absolute_error(y_test,y_pred)
```
