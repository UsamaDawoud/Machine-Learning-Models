```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
blobs = pd.read_csv('blobs.csv')
circles = pd.read_csv('circles.csv')
moons = pd.read_csv('moons.csv')
```

```python
sns.scatterplot(data=blobs, x='X1',y='X2');
```

```python
sns.scatterplot(data=circles, x='X1',y='X2');
```

```python
sns.scatterplot(data=moons, x='X1',y='X2');
```

```python
def model_apply(model,data):
    labels = model.fit_predict(data)
    sns.scatterplot(data=data,x='X1',y='X2',hue=labels);
```

```python
from sklearn.cluster import KMeans
model = KMeans(n_clusters=3)
model_apply(model,blobs)
```

```python
# it cluster the data set efficiently 
```

```python
model = KMeans(n_clusters=2)
model_apply(model,circles)
```

```python
# fill to cluster the data set in good way
```

```python
model= KMeans(n_clusters=2)
model_apply(model,moons)
```

```python
# fill to cluster the data set in good way
```

```python
from sklearn.cluster import AgglomerativeClustering
model=AgglomerativeClustering(n_clusters=3)
model_apply(model,blobs)
```

```python
# it cluster the data set efficiently 
```

```python
model=AgglomerativeClustering(n_clusters=2)
model_apply(model,circles)
```

```python
# fill to cluster the data set in good way
```

```python
model=AgglomerativeClustering(n_clusters=2)
model_apply(model,moons)
```

```python
# fill to cluster the data set in good way
```

```python
from sklearn.cluster import DBSCAN 
model=DBSCAN()
model_apply(model,blobs)
```

```python
# it cluster the data set efficiently 
```

```python
model=DBSCAN(eps=0.15)
model_apply(model,circles)
```

```python
# it cluster the data set efficiently 
```

```python
model=DBSCAN(eps=0.15)
model_apply(model,moons)
```
