```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data features for cancer tumor.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
# 30 columns !! 
```

```python
# scaling the data 
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_df = scaler.fit_transform(df)
```

```python
scaled_df
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca_df = pca.fit_transform(scaled_df)
pca_df
```

```python
pca_df.shape
```

```python
plt.scatter(pca_df[:,0], pca_df[:,1])
```

```python
# the data transformed from 30 columns to 2 columns only !!
# to figure out how much effects the results :
```

```python
from sklearn.datasets import load_breast_cancer
cancer_dictionary = load_breast_cancer()
cancer_dictionary['target']
```

```python
plt.scatter(pca_df[:,0], pca_df[:,1],c = cancer_dictionary['target'])
```

```python
df_comp = pd.DataFrame(pca.components_ , index = ['pc1','pc2'] , columns=df.columns)
df_comp
```

```python
plt.figure(figsize= (15, 4), dpi= 150)
sns.heatmap(df_comp , annot =True);
```

```python
pca.explained_variance_ratio_ 
```

```python
np.sum(pca.explained_variance_ratio_)
```

```python
# the model with remain 2 columns (out from 30 columns) can label 63 percent of the data 
```

```python
percent = []
for n in range(1 , 30):
    pca = PCA(n_components=n)
    pca_df = pca.fit_transform(scaled_df)
    percent.append(np.sum(pca.explained_variance_ratio_))
```

```python
plt.plot(range(1 , 30), percent);
```

```python
percent
```

```python
k = 0
for i in percent:
    k = k + 1
    if int(i*100 ) == 99 :
        print(k)
    
    
```

```python
# so when the value of (n_components) reach 17 then the model able to label 99 percent of data ! 
```
